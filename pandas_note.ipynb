## Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv('data/Flavors.csv')
```

```python
df
```

```python
agg_func = {
    'Flavor Rating' : ['mean', 'sum', 'max', 'min'],
    'Texture Rating': ['mean', 'sum', 'max', 'min'],
}
```

```python
group_by_frame = df.groupby('Base Flavor').agg(agg_func)
group_by_frame
```

```python
df.groupby(['Base Flavor', 'Liked']).agg(agg_func)
```

```python
df
```

```python
agg_funcs = {
    'Flavor': 'first',  # Take the first value in the group for the 'Flavor' column
    'Liked': lambda x: x.mode()[0] if not x.mode().empty else x.iloc[0],  # Take the most frequent value (mode) or the first if no mode
    'Flavor Rating': 'mean',  # Calculate the mean for 'Flavor Rating'
    'Texture Rating': 'mean',  # Calculate the mean for 'Texture Rating'
    'Total Rating': 'mean'  # Calculate the mean for 'Total Rating'
}
```

```python
group_by_frame = df.groupby('Base Flavor').agg(agg_funcs).reset_index()
group_by_frame
```

```python
pd.set_option('display.max.column', 24)
df.groupby('Base Flavor').describe()
```

## Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r'data/Customer Call List.xlsx')
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.drop(columns='Not_Useful_Column', inplace=True);
df
```

```python
df['Last_Name'].str.strip() # Remove leading and trailing characters.
```

```python
df["Last_Name"].str.strip("._/")
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df.replace('N/a', '', inplace=True);
df.fillna('', inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df.reset_index(drop=True, inplace=True);
df
```
